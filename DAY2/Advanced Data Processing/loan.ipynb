```python
Analyze customer financial data to classify:
Low / Medium / High risk

customer_id - Unique identifier for each customer
income - Annual income of the customer. Higher income → lower risk
loan_Amount - Amount the customer wants to borrow

Loan-to-Income Ratio=loan_amount/income

0.5 → High risk
0.3–0.5 → Medium
< 0.3 → Low

credit_score - Numeric score representing creditworthiness

Score Range	Risk Level
750+	Excellent
700–749	Good
600–699	Fair
<600	High Risk


defaults -  Number of past loan defaults / missed payments

Defaults	Risk
0	Low
1–2	Medium
≥3	Higher

Example Record Explained
customer_id: 101
income: 40000
loan_amount: 30000
credit_score: 580
defaults: 2
🔍 Analysis
Loan/Income = 0.75 → ❌ High
Credit Score = 580 → ❌ Poor
Defaults = 2 → ⚠️ Medium
👉 Final Risk: HIGH

write a automation to Combines all these fields
Assigns risk points
Converts score → category:
LOW
MEDIUM
HIGH








```

```python

import csv

def risk_score(income, loan, credit, defaults):
    score = 0

    # Income vs loan ratio
    ratio = loan / income
    if ratio > 0.5:
        score += 2
    elif ratio > 0.3:
        score += 1

    # Credit score
    if credit < 600:
        score += 2
    elif credit < 700:
        score += 1

    # Defaults
    if defaults >= 3:
        score += 2
    elif defaults > 0:
        score += 1

    return score

def classify(score):
    if score >= 5:
        return "HIGH"
    elif score >= 3:
        return "MEDIUM"
    else:
        return "LOW"

output = []

with open("loan_data.csv") as f:
    reader = csv.DictReader(f)
    for row in reader:
        score = risk_score(
            int(row["income"]),
            int(row["loan_amount"]),
            int(row["credit_score"]),
            int(row["defaults"])
        )
        category = classify(score)
        output.append([row["customer_id"], score, category])

# Save output
with open("risk_output.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["customer_id","risk_score","risk_category"])
    writer.writerows(output)

print("Risk profiling completed. Output saved.")
```
