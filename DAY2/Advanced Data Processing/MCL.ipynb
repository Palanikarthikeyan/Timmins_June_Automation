```python
Infrastructure: Multi-Cloud Latency & Cost Optimizer
Complexity: Merging real-time performance logs with a static pricing CSV to find the best cost-to-performance ratio.

Source A (JSON API): Real-time latency from servers in different regions.

Source B (CSV File): Hourly cost per instance type.

Goal: Identify servers that are "Expensive but Slow" and suggest a migration.
```

```python
import pandas as pd

def run_infrastructure_analysis():
    # Load Data Source 1 (JSON)
    # Similar to parsing system logs or alarm logs [cite: 10, 14]
    df_latency = pd.read_json('latency_logs.json')
    
    # Load Data Source 2 (CSV)
    # Similar to processing multi-file CSV batches 
    df_pricing = pd.read_csv('pricing_table.csv')

    # Merge Dataframes (Relational Join on 'id')
    # This identifies the specific machine/server characteristics [cite: 7, 8]
    combined_df = pd.merge(df_latency, df_pricing, on="id")

    # Add Logic: Efficiency Score (Latency * Cost)
    # Lower is better. High scores indicate "Expensive and Slow"
    combined_df['efficiency_index'] = combined_df['latency_ms'] * combined_df['cost_hr']

    # Identify "Spikes" or Outliers [cite: 13, 14]
    # Filter for latency > 100ms and cost > 0.80
    high_risk = combined_df[(combined_df['latency_ms'] > 100) & (combined_df['cost_hr'] > 0.80)]

    return combined_df, high_risk

# Execute
full_report, alerts = run_infrastructure_analysis()

print("--- Full Consolidated Infrastructure Report ---")
print(full_report)

print("\n--- Critical Efficiency Alerts ---")
if not alerts.empty:
    print(alerts)
else:
    print("No efficiency issues detected.")
```
