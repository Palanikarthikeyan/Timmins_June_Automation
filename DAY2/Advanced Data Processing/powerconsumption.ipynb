```python
use case: Power Consumption Analyzer
Analyze device power logs.
Tasks
Detect abnormal consumption patterns
Compare against baseline
Identify faulty components

Instead of giving power directly, we compute it:

P=V×I
V = voltage
I = current
P = power (Watts)


baseline = {
    "CPU": 1.0,
    "GPU": 0.7,
    "MODEM": 0.4
}

If power > baseline × threshold (e.g., 1.5) → anomaly  
```

```python
import csv
from collections import defaultdict

file = "power_logs.csv"

baseline = {
    "CPU": 1.0,
    "GPU": 0.7,
    "MODEM": 0.4
}

THRESHOLD = 1.5

anomalies = defaultdict(int)
total_records = 0

with open(file) as f:
    reader = csv.DictReader(f)
    
    for row in reader:
        total_records += 1
        
        component = row["component"]
        voltage = float(row["voltage"])
        current = float(row["current"])
        
        power = voltage * current
        
        # Check anomaly
        if power > baseline[component] * THRESHOLD:
            anomalies[component] += 1

# Identify faulty components
faulty = [comp for comp, count in anomalies.items() if count >= 2]

# Output
print("===== Power Consumption Report =====")
print(f"Total Records: {total_records}")

print("\nAnomalies Detected:")
for comp, count in anomalies.items():
    print(f"{comp}: {count}")

print("\nFaulty Components:")
for comp in faulty:
    print(comp)
```
