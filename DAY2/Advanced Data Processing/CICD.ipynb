```python
CI/CD Pipeline Failure Analyzer (Advanced Lab)
Problem Statement
Your CI/CD system generates thousands of build logs daily.
You need to:
Parse logs
Identify failure patterns
Group failures by root cause
Generate trend summary
```

```python
import re
from collections import defaultdict

log_file = "build_logs.txt"

failure_counts = defaultdict(int)
total_builds = 0
failed_builds = 0
success_builds = 0

pattern = re.compile(
    r'(?P<time>.*?) BUILD_ID=(?P<id>\d+) STATUS=(?P<status>\w+)(?: ERROR=(?P<error>\w+))?'
)

with open(log_file) as f:
    for line in f:
        match = pattern.search(line)
        if not match:
            continue

        total_builds += 1
        status = match.group("status")

        if status == "SUCCESS":
            success_builds += 1
        else:
            failed_builds += 1
            error = match.group("error")
            failure_counts[error] += 1

# Trend Analysis
failure_rate = (failed_builds / total_builds) * 100

most_common_failure = max(failure_counts, key=failure_counts.get)

# Output
print("===== CI/CD Failure Report =====")
print(f"Total Builds: {total_builds}")
print(f"Success Builds: {success_builds}")
print(f"Failed Builds: {failed_builds}")
print(f"Failure Rate: {failure_rate:.2f}%")

print("\nFailure Breakdown:")
for error, count in failure_counts.items():
    print(f"{error}: {count}")

print(f"\nMost Common Failure: {most_common_failure}")
```
