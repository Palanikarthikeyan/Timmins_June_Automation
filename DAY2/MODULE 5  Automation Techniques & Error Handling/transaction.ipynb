```python
UseCase: Fraudulent Transaction Pattern Detection
Industry: Fintech / Banking / Cyber Security

Problem Statement

A payment gateway processes thousands of transactions per minute. 
To prevent financial loss, the system must monitor transaction logs in real-time 
to identify suspicious activity based on velocity and volume.

You need to: 


Parse transaction logs for amount and status. 

Identify accounts with more than 3 failed attempts in a 10-minute window. 

Detect "High Value" spikes (transactions > rupess 10,000). 


Generate a Fraud Alert Summary for the security team. 
```

```python
transaction_logs = [
    {"user": "USER_A", "amount": 500, "status": "SUCCESS", "time": "12:00"},
    {"user": "USER_B", "amount": 12000, "status": "SUCCESS", "time": "12:01"},
    {"user": "USER_C", "amount": 50, "status": "FAILED", "time": "12:02"},
    {"user": "USER_C", "amount": 45, "status": "FAILED", "time": "12:03"},
    {"user": "USER_C", "amount": 60, "status": "FAILED", "time": "12:05"},
    {"user": "USER_C", "amount": 55, "status": "FAILED", "time": "12:08"},
    {"user": "USER_D", "amount": 15000, "status": "SUCCESS", "time": "12:10"}
]
```

```python
from collections import defaultdict

def analyze_transactions(logs, high_value_threshold=10000, failure_threshold=3):
    fraud_report = {
        "high_value_alerts": [],
        "suspicious_users": {},
        "total_processed": len(logs)
    }
    
    # Tracking failures per user
    failure_counts = defaultdict(int)

    for entry in logs:
        user = entry["user"]
        amount = entry["amount"]
        status = entry["status"]

        # 1. Detect High-Value Spikes
        if amount > high_value_threshold:
            fraud_report["high_value_alerts"].append({
                "user": user, 
                "amount": amount, 
                "time": entry["time"]
            })

        # 2. Count Failed Attempts
        if status == "FAILED":
            failure_counts[user] += 1
            if failure_counts[user] > failure_threshold:
                fraud_report["suspicious_users"][user] = failure_counts[user]

    return fraud_report

# Sample Input Data based on the Fintech Use Case
transaction_logs = [
    {"user": "USER_A", "amount": 500, "status": "SUCCESS", "time": "12:00"},
    {"user": "USER_B", "amount": 12000, "status": "SUCCESS", "time": "12:01"},
    {"user": "USER_C", "amount": 50, "status": "FAILED", "time": "12:02"},
    {"user": "USER_C", "amount": 45, "status": "FAILED", "time": "12:03"},
    {"user": "USER_C", "amount": 60, "status": "FAILED", "time": "12:05"},
    {"user": "USER_C", "amount": 55, "status": "FAILED", "time": "12:08"},
    {"user": "USER_D", "amount": 15000, "status": "SUCCESS", "time": "12:10"}
]

# Run Analysis
report = analyze_transactions(transaction_logs)

# --- Generate Summary Report ---
print("--- FRAUD ANALYSIS REPORT ---")
print(f"Total Transactions Parsed: {report['total_processed']}")

print("\n[HIGH VALUE ALERTS]")
for alert in report['high_value_alerts']:
    print(f"ALERT: User {alert['user']} transacted Rupees {alert['amount']} at {alert['time']}")

print("\n[SUSPICIOUS ACTIVITY]")
if not report['suspicious_users']:
    print("No velocity flags detected.")
else:
    for user, count in report['suspicious_users'].items():
        print(f"FLAG: {user} reached {count} failed attempts. Automated Block Triggered.")
```
