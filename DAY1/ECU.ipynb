```python
Industry: Automobile / Embedded Systems / IoT 

Problem Statement

An Electronic Control Unit (ECU) in a vehicle must send a "Heartbeat" signal every 100ms. If the signal is missed three times in a row, the system must trigger a safety "Limp Mode".

You need to:

Monitor incoming timestamps of sensor signals.

Detect gaps in communication exceeding 300ms.

Flag the specific module failing the heartbeat check.
```

```python
# Signal intervals in milliseconds
heartbeat_timestamps = [100, 200, 550, 650, 750] # Gap between 200 and 550 is 350ms

def monitor_heartbeat(data, threshold=300):
    for i in range(1, len(data)):
        gap = data[i] - data[i-1]
        if gap > threshold:
            return f"ALERT: Heartbeat Missing! Gap: {gap}ms. Triggering Safety Protocol."
    return "Status: All Systems Nominal"

print(monitor_heartbeat(heartbeat_timestamps))
```
